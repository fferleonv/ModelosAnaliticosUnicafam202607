```python
#La cuenta de un restaurante es $85,000. Calcula e imprime: la propina del 15%, el total a pagar (cuenta + propina), y cuánto paga cada persona si son 4.

cuenta = 85000
porcentaje_propina = 0.15
personas = 4

propina = cuenta * porcentaje_propina
total_a_pagar = cuenta + propina
pago_por_persona = total_a_pagar / personas

print(f"Propina (15%): ${propina}")
print(f"Total a pagar: ${total_a_pagar}")
print(f"Cada persona paga: ${pago_por_persona}")
```

```python
# Un parque de diversiones cobra $15,000 a adultos (18-60 años) y $8,000 a niños o adultos mayores. El precio se duplica los fines de semana. Dado edad = 25 y fin_de_semana = True, calcula el precio a pagar usando operadores lógicos.

edad = 25
fin_de_semana = True
es_adulto = (edad >= 18) and (edad <= 60)
precio_base = 15000 if es_adulto else 8000
precio_final = precio_base * 2 if fin_de_semana else precio_base
print(f"Edad: {edad} años")
print(f"¿Es fin de semana?: {fin_de_semana}")
print(f"Total a pagar: ${precio_final}")
```

```python
#Dado peso = 70 (kg) y altura = 1.75 (m), calcula el IMC (peso / altura²) e imprime la categoría: Bajo peso (<18.5), Normal (18.5–24.9), Sobrepeso (25–29.9), Obesidad (≥30).

peso = 70
altura = 1.75
imc = peso / (altura ** 2)
if imc < 18.5:
    categoria = "Bajo peso"
elif imc <= 24.9:
    categoria = "Normal"
elif imc <= 29.9:
    categoria = "Sobrepeso"
else:
    categoria = "Obesidad"
print(f"Tu IMC es: {imc:.2f}")
print(f"Categoría: {categoria}")
```

```python
#Un sistema requiere que el usuario tenga más de 18 años Y que su contraseña tenga más de 8 caracteres. Dado edad = 20 y password = "abc123", imprime si el acceso es permitido o cuál requisito falla.


edad = 20
password = "abc123"
cumple_edad = edad > 18
cumple_password = len(password) > 8
if cumple_edad and cumple_password:
    print("Acceso permitido")
else:
    print("Acceso denegado. Detalles del fallo:")
    if not cumple_edad:
        print("- El usuario debe tener más de 18 años.")
    if not cumple_password:
        print(f"- La contraseña debe tener más de 8 caracteres (actualmente tiene {len(password)}).")
```

```python
#Calculadora de descuento
#Una tienda aplica descuentos según el monto: 0% si es menos de $50,000; 10% entre $50,000 y $99,999; 20% entre $100,000 y $199,999; 30% si es $200,000 o más. Dado monto = 120000, calcula e imprime el descuento y el total a pagar.
monto = 120000
if monto < 50000:
    porcentaje_descuento = 0.0
elif monto <= 99999:
    porcentaje_descuento = 0.10
elif monto <= 199999:
    porcentaje_descuento = 0.20
else:
    porcentaje_descuento = 0.30
valor_descuento = monto * porcentaje_descuento
total_a_pagar = monto - valor_descuento

print(f"Monto de la compra: ${monto:,}")
print(f"Descuento aplicado: {porcentaje_descuento * 100:.0f}% (${valor_descuento:,})")
print(f"Total final a pagar: ${total_a_pagar:,}")
```

```python
#Para un número n, imprime: "FizzBuzz" si es múltiplo de 3 y de 5; "Fizz" si es múltiplo solo de 3; "Buzz" si es múltiplo solo de 5; el número si no cumple ninguna. Prueba con n = 15, n = 9, n = 10 y n = 7.
numeros_a_probar = [15, 9, 10, 7]
for n in numeros_a_probar:
    if n % 3 == 0 and n % 5 == 0:
        resultado = "FizzBuzz"
    elif n % 3 == 0:
        resultado = "Fizz"
    elif n % 5 == 0:
        resultado = "Buzz"
    else:
        resultado = str(n)
    print(f"Para n = {n} 👉 {resultado}")
```

```python
#Usando un ciclo for, suma todos los números del 1 al 100 e imprime el resultado. Luego suma solo los números pares del mismo rango.

suma_total = 0
suma_pares = 0
for i in range(1, 101):
        suma_total += i
if i % 2 == 0:
        suma_pares += i
print(f"La suma de todos los números del 1 al 100 es: {suma_total}")
print(f"La suma de solo los números pares del 1 al 100 es: {suma_pares}")
```

```python
#Dada precios = [12000, 45000, 8500, 67000, 3200], usa un ciclo para: (1) imprimir cada precio con formato Item 1: $12,000, (2) imprimir al final cuántos artículos cuestan más de $10,000.
precios = [12000, 45000, 8500, 67000, 3200]
articulos_caros = 0
for i, precio in enumerate(precios):
    print(f"Item {i + 1}: ${precio:,}")
    if precio > 10000:
        articulos_caros += 1
print("---")
print(f"Cantidad de artículos que cuestan más de $10,000: {articulos_caros}")

```

```python
#Un banco tiene saldo = 1000. Cada mes retiras 150 y se cobran 20 de comisión. Usa un ciclo while para simular cuántos meses pasan antes de que el saldo llegue a cero o menos. Imprime el saldo al final de cada mes.

saldo = 1000
mes = 0
print(f"Saldo inicial: ${saldo}\n")
while saldo > 0:
    mes += 1
    saldo -= (150 + 20)
    print(f"Mes {mes}: Saldo restante = ${saldo}")
print("---")
print(f"El saldo se agotó. Pasaron {mes} meses en total.")
```
