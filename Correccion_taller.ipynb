TALLER ACTUALIZACION


```python
# DICCIONARIO CON DATOS BASICOS
estudiante={
    "Nombre":"Jeniffer",
    "Apellido":"Leon",
    "Ciudad":"Bogota",
    "Hijos":"No",
    "Carrera":"Administracion de empresas"
}

```

```python
# Construccion del menu
print("----Bienvenido a tu programa de actualizacion de datos----")
print("Para comenzar elige la opciòn deseada")
print("1.Agregar un dato nuevo")
print("2.Actualizar un dato")
print("3.Eliminar un dato con del")
print("4.Eliminar un dato con pop")
print("5.Consultar un dato")
print("6.Imprimir diccionario completo")
seleccion=int(input("Selecciona la opcion deseada:"))

```

```python
# Agregar un dato nuevo
if seleccion==1:
  clave=input("Ingrese la clave deseada:")
  valor=input("Ingrese valor deseado")

  #Agregar elemento nuevo

  estudiante[clave]=valor
  print("El valor se agrego exitosamente")
  print(estudiante)

# Actualizar
elif seleccion==2:
  clave=input("Ingrese la clave actualizar")
  estudiante[clave]=input("Ingrese valor nuevo")
  print("El dato fue actualizado con exito")
  print(estudiante)

#Eliminar
elif seleccion==3:
  clave=input("Ingrese la clave a eliminar")
  del estudiante[clave]
  print("El Dato fue eliminado exitosamente")
  print(estudiante)

# eEliminar con pop
elif seleccion==4:
  clave=input("ingrese clave a eliminar")
  estudiante.pop(clave)
  print("El dato fue eliminado exitosamente")
  print(estudiante)

# Consultar dato
elif seleccion==5:
  clave=input("Ingrese clave a consultar")
  valor=estudiante.get(clave,"no disponible")
  print(f"{clave}:{valor}")


elif seleccion==6:
  print("perfil completo")
  print(estudiante)

else:
  print("Opcion invalida")





```
