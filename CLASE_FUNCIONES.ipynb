```python
# PRESENTACION CON EDAD
def presentacion(nombre,edad=47):
  print(f"Hola mi nombre es ¨{nombre} y tengo {edad} años")
presentacion("jeniffer leon")

```

```python
# GENERA CALCULADORA DE OPERACIONES
def operaciones (a,b):
  suma=a+b
  resta=a-b
  mult=a*b
  division=a/b
  return suma,resta,mult,division
result_suma, result_resta,result_mult,result_division=operaciones(8,5)
print(f"suma: {result_suma}, resta {result_resta}, Multiplicacion {result_mult}, Division {result_division}")
```

```python
# ESCRIBE UNA FUNCION QUE RECIBA LA EDAD DE UNA PERSONA COMO PARAMETRO Y DETERMINE SI LA PERSONA ES MAYOR DE EDAD O MENOR
def evaluar_edad(edad):
  if edad >= 18:
    return "Es mayor de edad"
  else:
    return (" Es menor de edad")

resultado1 = evaluar_edad(20)
print(f"Para 20 años: {resultado1}")
resultado2 = evaluar_edad(15)
print(f"para 15 años: {resultado2}")
resultado3 = evaluar_edad(18)
print(f"para 18 años: {resultado3}")


```

## EJERCICIO 4

```python
# Calcular IMC
def imc(peso, altura):
  valor_imc = peso /(altura**2)
  if valor_imc<18.5:
   return "Bajo peso"
  elif valor_imc<=24.9:
   return "Normal"
  elif valor_imc<=29.9:
   return " Sobrepeso"
  else:
   return "Obesidad"

print(f"Resultado 1: {imc(50, 1.70)}")   # Peso bajo
print(f"Resultado 2: {imc(68, 1.70)}")   # Normal
print(f"Resultado 3: {imc(80, 1.70)}")   # Sobrepeso
print(f"Resultado 4: {imc(95, 1.70)}")   # Obesidad
```
