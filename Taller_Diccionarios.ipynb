```python
# Acceder a la info
# 2. Impresión del menú de opciones
print("     MENÚ DE MI DICCIONARIO - PYTHON     ")
print("1. Agregar un dato nuevo")
print("2. Actualizar un dato existente")
print("3. Eliminar un dato con del")
print("4. Eliminar un dato con .pop()")
print("5. Consultar un dato con .get()")
print("6. Ver el perfil completo")
```

```python
# CLAVE
opcion = input("Por favor, elija una opción (1-6): ")
print("-" * 40)
```

```python
# CONDICIONALES

if opcion == "1":
    # Opción 1: Agregar un dato nuevo (Asignación directa con corchetes)
    clave = input("Ingrese la nueva clave (ej. pasatiempo, ciudad): ")
    valor = input(f"Ingrese el valor para '{clave}': ")
    perfil[clave] = valor  # Asignación directa
    print("\n[ÉXITO] Dato agregado correctamente.")
    print("Perfil actualizado:", perfil)

elif opcion == "2":
    # Opción 2: Actualizar un dato existente (Verificación previa con 'in')
    clave = input("Ingrese la clave que desea actualizar: ")
    if clave in perfil:
        nuevo_valor = input(f"Ingrese el nuevo valor para '{clave}': ")
        perfil[clave] = nuevo_valor
        print("\n[ÉXITO] Dato actualizado correctamente.")
        print("Perfil actualizado:", perfil)
    else:
        print(f"\n[ALERTA] La clave '{clave}' no existe en el perfil.")
elif opcion == "3":
    # Opción 3: Eliminar un dato con del (Sin conservar valor)
    clave = input("Ingrese la clave que desea eliminar con 'del': ")
    if clave in perfil:
        del perfil[clave]  # Palabra reservada del
        print(f"\n[ÉXITO] La clave '{clave}' fue eliminada permanentemente.")
        print("Perfil actualizado:", perfil)
    else:
        print(f"\n[ALERTA] La clave '{clave}' no existe, no se pudo eliminar.")
elif opcion == "4":
    # Opción 4: Eliminar un dato con .pop() (Conservando el valor eliminado)
    clave = input("Ingrese la clave que desea eliminar con '.pop()': ")
    if clave in perfil:
        valor_eliminado = perfil.pop(clave)  # Método .pop() extrae y retorna el valor
        print(f"\n[ÉXITO] Clave '{clave}' eliminada.")
        print(f"El valor que se eliminó y se guardó en la variable fue: '{valor_eliminado}'")
        print("Perfil actualizado:", perfil)
    else:
        print(f"\n[ALERTA] La clave '{clave}' no existe, no se pudo eliminar.")
elif opcion == "5":
    # Opción 5: Consultar un dato seguro con .get()
    clave = input("Ingrese la clave que desea consultar con '.get()': ")
    # .get(clave, valor_por_defecto) evita que el programa falle si la clave no existe
    resultado = perfil.get(clave, "No disponible")
    print(f"\nResultado de la consulta para '{clave}': {resultado}")

elif opcion == "6":
    # Opción 6: Ver el perfil completo
    print("\nVisualizando el Perfil Completo:")
    print(perfil)
else:
    # Manejo de opciones inválidas
    print("\n[ERROR] Opción no válida. Debe introducir un número del 1 al 6.")
```
