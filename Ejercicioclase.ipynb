## OPERADORES MATEMATICOS

```python
# SUMA
a=10
b=3
suma=a+b
print(a+b)
print(suma)
```

```python
# RESTA
resta=a-b
print(a-b)
print(resta)
```

```python
# MULTIPLICACION
multiplicacion=a*b
print(a*b)
print(multiplicacion)
```

```python
# DIVISION
division=a/b
print(a/b)
print(division)
```

```python
# MODULO
print(a%b)
```

```python
# LA EXPONENCIACION
potencia=a**b
print(potencia)
```

```python
# DIVISION ENTERA
print(a//b)
```
